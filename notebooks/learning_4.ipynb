```c++17
#include <iostream>
#include <string>
#include <cmath>


std::string decimalToBinary(double num, int precision) {
    // 处理整数部分
    int integerPart = static_cast<int>(num);
    double fractionalPart = num - integerPart;
    std::string binaryResult = "";

    // 转换整数部分到二进制
    while (integerPart > 0) {
        binaryResult = std::to_string(integerPart % 2) + binaryResult;
        integerPart /= 2;
    }

    // 如果原始数为小数，添加小数点
    if (fractionalPart > 0) {
        binaryResult += ".";
    }

    // 转换小数部分到二进制
    while (precision-- > 0 && fractionalPart > 0) {
        fractionalPart *= 2;
        int binaryBit = static_cast<int>(fractionalPart);
        binaryResult += std::to_string(binaryBit);
        fractionalPart -= binaryBit;
    }

    return binaryResult;
}
```

```c++17
int main() {
    double num = 12.375;
    int precision = 5;  // 小数位精度
    string mybinary = decimalToBinary(num, precision);
    cout << "十进制数 " << (num) << " 的二进制表示为: " << (mybinary) << endl;
    return 0;
}
```

■ 布尔值转换（boolean conversions） 算术型、枚举型、指针型以及成员的指针型的右值可 以转换为 bool 型的右值。0、空指针及空成员指针的 值转换为 false，其他值都转换为 true。

```c++17
int a = 0;
int b = 42;
float c = 0.0;
float d = 3.14;
int* e = nullptr;
int f = 1;
int* g = &f;
enum Color { Red, Green, Blue };
Color color = Green;

bool ba = (bool)a; // false，因为 a 为 0
bool bb = (bool)b; // true，因为 b 非 0
bool bc = (bool)c; // false，因为 c 为 0.0
bool bd = (bool)d; // true，因为 d 非 0.0
bool be = (bool)e; // false，因为 e 是空指针
bool bg = (bool)g; // true，因为 g 是非空指针
bool bcolor = (bool)color; // true，因为 color 非 0（Green 实际值为 1）

cout << std::boolalpha;
cout << "a: " << ba << '\n';
cout << "b: " << bb << '\n';
cout << "c: " << bc << '\n';
cout << "d: " << bd << '\n';
cout << "e: " << be << '\n';
cout << "g: " << bg << '\n';
cout << "color: " << bcolor << '\n';
```

## 枚举体

```c++17
// 显示所选动物的叫声

int main()
{
	enum animal { Dog, Cat, Monkey, Invalid };
	int type;

	do {
		cout << "0...狗 1...猫 2...猴 3...结束：";
		cin >> type;
	} while (type < Dog || type > Invalid);

	if (type != Invalid) {
		animal selected = static_cast<animal>(type);
		switch (selected) {
		 case Dog	 : cout << "汪汪!!\n";	break;
		 case Cat	 : cout << "喵喵!!\n";	break;
		 case Monkey : cout << "唧唧!!\n";	break;
		}
	}
}
main()
```
